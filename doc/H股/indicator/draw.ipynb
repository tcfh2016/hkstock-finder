```python
import akshare as ak
import numpy as np
import matplotlib.pyplot as plt

stock_hk_valuation_pe_df = ak.stock_hk_valuation_baidu(symbol="06098", indicator="总市值", period="全部")
stock_hk_valuation_pe_df = stock_hk_valuation_pe_df.set_index("date")
stock_hk_valuation_pe_df.columns = ["总市值"]
ax = stock_hk_valuation_pe_df["总市值"].plot(figsize=(10, 5))

major_ticks = np.arange(0, 3001, 500)
minor_ticks = np.arange(0, 3001, 100)
ax.set_yticks(minor_ticks, minor=True)
ax.set_yticks(major_ticks)
ax.grid(which='both')
ax.grid(which='minor', alpha=0.5)
ax.grid(which='major', alpha=0.9)
ax.grid('on', which='minor')
ax.grid('on', which='major')
```
